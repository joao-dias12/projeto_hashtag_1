### Passo 1 - Importar Arquivos e Bibliotecas

```python
#Importing libraries
import pandas as pd
```

```python
#importing data base
try:
    emails = pd.read_excel(r'Bases de Dados/Emails.xlsx')
    lojas = pd.read_csv(r'Bases de Dados/Lojas.csv', encoding='latin1', sep=';')
    vendas = pd.read_excel(r'Bases de Dados/Vendas.xlsx')


except Exception as err:
    print(f'A classe do erro foi {err.__class__} e o contexto  foi {err.__context__}')
else:
    print('A base de dados foi importada com sucesso')
display(emails)
```

Criar uma tabela para cada loja

```python
#merging table of sells with stores
vendas = vendas.merge(lojas, on='ID Loja')
display(vendas)


```

```python
#Criando um dicionário para armazenar cada loja com suas vendas
tabela_lojas = {}
for loja in lojas['Loja']:
    tabela_lojas[loja] = vendas.loc[vendas['Loja'] == loja, :]

display(tabela_lojas['Iguatemi Esplanada'])
```

### Passo 2 - Definir Criar uma Tabela para cada Loja e Definir o dia do Indicador

```python
# UM DOS INDICADORES ENVIADOS NO EMAIL SERA O VALOR DO DIA, LOGO O ULTIMO DIA DA PLANILHA , como identificar o ultimo dia ?
dia_do_indicador = vendas['Data'].max() # Dessa forma sempre vai pegar a maior data, a data mais recente

print(dia_do_indicador)
```

### Passo 3 - Salvar a planilha na pasta de backup

```python
#identificar se a pasta ja existe
import pathlib
import os
from pprint import pprint
pasta_backup = pathlib.Path(r'Backup Arquivos Lojas')
arquivos_lista_backup = pasta_backup.iterdir()
lista_nome_backup = [arquivo.name for arquivo in arquivos_lista_backup] #lita com nome das pastas 

print(lista_nome_backup)

for loja, tabela in tabela_lojas.items():
    if loja not in lista_nome_backup: # Dessa forma , caso não surja uma loja nova, ele não vai criar uma nova pasta das lojas que já existem 
        os.makedirs(f'{pasta_backup}/{loja}')
        
    tabela.to_excel(f'{pasta_backup}/{loja}/{dia_do_indicador.day}_{dia_do_indicador.month}_{loja}.xlsx') #criando arquivo excel


```

### Passo 4 - Calcular o indicador para 1 loja

```python
tabela_ano = tabela_lojas['Iguatemi Esplanada'] #Transformando em tabela
tabela_ultimo_dia = tabela_ano[tabela_ano['Data'] == dia_do_indicador] #filtrando o campo "data" pelo ultimo idia registrado

# Faturamento 
faturamento_ano = tabela_ano['Valor Final'].sum() #recebendo a soma da coluna
meta_fat_ano = 1650000
ultima_dia_faturamento = tabela_ultimo_dia['Valor Final'].sum()
meta_fat_dia = 1000

#diversidade de Produtos
qtde_produtos_ano = len(tabela_ano['Produto'].unique()) # Estou pegando os valores unicos de uma coluna, transformando em uma lista e calculando o tamnho da lista
qtde_produtos_dia = len(tabela_ultimo_dia['Produto'].unique())
meta_qtde_prod_dia = 4
meta_qtde_prod_ano = 120
meta_ticket_medio = 500



#ticket médio
venda_group_cod = tabela_ano.groupby('Código Venda').sum() #Agrupando a tabela utilizando a coluna "Código de Vendas" e somando as outras colunas
ticket_medio_ano = venda_group_cod['Valor Final'].mean() # Calculando a média do valor final

venda_group_cod_dia = tabela_ultimo_dia.groupby('Código Venda').sum()
ticket_medio_dia = venda_group_cod_dia['Valor Final'].mean()
 
display(venda_group_cod , ticket_medio_ano)
    
```
